### casos sencillos

```python
# caso sencillo

# definimos dos arreglos
years = [1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
population = [22.6, 39.7, 50.5, 63.2, 75.8, 90.6, 101.5, 117.0]

# (datos eje x, datos eje y, color, marcador, tipo de línea)
plt.plot(years, population, color='blue',  marker='o', linestyle='solid')
# título y tamaño de letra
plt.title('Popuation Mexico', fontsize=20)
# etiqueta eje x
plt.xlabel('year', fontsize=16)
# etiqueta eje y
plt.ylabel('millions', fontsize=16)
# muestra la gráfica
plt.show()
```

```python
# arreglo (inicio, fin, paso)
t = np.arange(0, 5, 0.2)

# rojo --, azul cuadrados, triángulos verdes 
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]

# suma de valores
total_error = [x+y for x,y in zip(variance, bias_squared)]
xs = [i for i,_ in enumerate(variance)]

plt.plot(xs, variance,     'g-' , label = 'variance')           
plt.plot(xs, bias_squared, 'r-.', label = 'bias$^2$')           
plt.plot(xs, total_error,  'b:' , label = 'total error')

# posición de las leyendas
plt.legend(loc = 'upper center')                            
plt.xlabel('model')
plt.title('Bias-variance tradeoff')
plt.show()
```

### gráfica a partir de una función

```python
# defino una función, en este caso $e^{-t} cos(2/pi t)$
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.2)
t2 = np.arange(0.0, 5.0, 0.1)

# gráfica principal
plt.figure(1)
# subgráficas
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'g--')

plt.show()
```

### diferentes escalas

```python
from matplotlib.ticker import NullFormatter  # útil para escala `logit` 

np.random.seed(19680801)

# se crean algunos datos en el intervalo (0,1) 
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure(1)

# escala lineal
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
# escala logarítmica
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
# escala logarítmica simétrica 
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)
# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

# Formato que evita la congestión en los ejes en caso de habr muchas etiquetas.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Ajuste de colocación de las gráficas
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5, wspace=0.5)
plt.show()
```

### histogramas

```python
# importamos un contador
from collections import Counter

calificaciones  = [80, 73, 75, 82, 64, 95, 86, 92, 0, 85, 71, 24, 100, 78, 81, 0, 0 ]
# rango de 10
rango = lambda nota: nota // 10*10

histograma = Counter(rango(nota) for nota in calificaciones)
plt.bar([x for x in histograma.keys()], histograma.values(), width = 5) 

plt.xticks([10*i for i in range(11)])
plt.xlabel('calificacion')
plt.ylabel('numero de estudiantes')
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# histograma de los datos (datos, bins, patches
plt.hist(x, 50, normed=1, facecolor='green')

plt.xlabel('IQ')
plt.ylabel('Probabilidad')
plt.title('Histograma de IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

### gráfica de flujo

```python
# se define una malla
Y, X = np.mgrid[-3:3:100j, -3:3:100j]

# se define el flujo
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

fig0, ax0 = plt.subplots()
strm = ax0.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
# barra de colores
fig0.colorbar(strm.lines)

plt.show()
```

### gráficas en 3D

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')

# párametro theta
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z, label='curva parametrica')
#ax.legend()
mpl.rcParams['legend.fontsize'] = 10

plt.show()
```

```python
# función auxiliar para crear un arrego de tamaño n distribuido uniformemente en (vmin, vmax).
    
def randrange(n, vmin, vmax):
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 50

# grafica n puntos para cada conjunto con estilo y rango predeterminado 
# en una caja definida en x [23, 32], y [0, 100], z [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('eje X')
ax.set_ylabel('eje Y')
ax.set_zlabel('eje Z')

plt.show()
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')

# malla y función
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# gráfica de la superficie
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# configuración del eje z
ax.set_zlim(-2, 2)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# agrega barra de colores
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
# define la malla
a,b = -1,1
n = 50
x,y = np.mgrid[a:b:(1j*n),a:b:(1j*n)]
# variable compleja
z = x + 1j*y
# función compleja
w = z**2

fig = plt.figure(figsize=(16,8))

# grafica la parte real
ax_real = fig.add_subplot(1,2,1,projection='3d')
ax_real.plot_surface(z.real, z.imag, w.real,
                    rstride=1, cstride=1, cmap=cm.jet)

# grafica la parte imaginaria
ax_imag = fig.add_subplot(1,2,2,projection='3d')
ax_imag.plot_surface(z.real, z.imag, w.imag,
                    rstride=1, cstride=1, cmap=cm.jet)
plt.show()
```
